## Training a Car Price Prediction Model

I will use some regression algorithm to train a car price prediction model. So let’s split the data into training and test sets and use the decision tree regression algorithm to train the model

## Feature Selection

```python
predict = "price"
data = data[["symboling", "wheelbase", "carlength",
             "carwidth", "carheight", "curbweight",
             "enginesize", "boreratio", "stroke",
             "compressionratio", "horsepower", "peakrpm",
             "citympg", "highwaympg", "price"]]
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Training a machine learning model

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

## Model Score

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model.score(xtest, predictions) 
```

```python
model.score(xtest, ytest)
```

```python
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(ytest, predictions)
print(f"Mean Absolute Error (MAE): {mae}")

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(ytest, predictions)
print(f"Mean Squared Error (MSE): {mse}")
# Calculate R-squared (R²) score
r_squared = r2_score(ytest, predictions)
print(f"R-squared (R²) Score: {r_squared}")
```

## The model gives 100% accuracy on the test set, which is excellent.
